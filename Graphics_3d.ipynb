```
import math
import matplotlib.pyplot as plt

def rotate_point_z(x, y, z, angle_deg):
    """
    Rotate a point around the Z-axis by a given angle in degrees.
    """
    angle_rad = math.radians(angle_deg)
    cos_theta = math.cos(angle_rad)
    sin_theta = math.sin(angle_rad)

    x_new = x * cos_theta - y * sin_theta
    y_new = x * sin_theta + y * cos_theta
    return (x_new, y_new, z)

def rotate_triangle_z(triangle, angle_deg):
    """
    Rotate all points of the triangle around the Z-axis.
    """
    return [rotate_point_z(x, y, z, angle_deg) for x, y, z in triangle]

def plot_triangles(original, rotated):
    """
    Plot the original and rotated triangles in 3D using matplotlib.
    """
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')

    x0, y0, z0 = zip(*original)
    x0 += (x0[0],)
    y0 += (y0[0],)
    z0 += (z0[0],)
    ax.plot(x0, y0, z0, color='blue', label='Original Triangle')
    ax.scatter(x0[:-1], y0[:-1], z0[:-1], color='blue')

    x1, y1, z1 = zip(*rotated)
    x1 += (x1[0],)
    y1 += (y1[0],)
    z1 += (z1[0],)
    ax.plot(x1, y1, z1, color='red', label='Rotated Triangle')
    ax.scatter(x1[:-1], y1[:-1], z1[:-1], color='red')

    ax.set_xlabel("X Axis")
    ax.set_ylabel("Y Axis")
    ax.set_zlabel("Z Axis")
    ax.set_title("Rotation of Triangle around Z-axis")
    ax.legend()
    ax.grid(True)
    plt.show()

def main():
    print("3D Triangle Rotation around Z-axis")

    triangle = []
    for i in range(3):
        print(f"\nEnter coordinates of point {i+1}:")
        x = float(input("  X: "))
        y = float(input("  Y: "))
        z = float(input("  Z: "))
        triangle.append((x, y, z))

    angle = float(input("\nEnter rotation angle (in degrees): "))
    rotated = rotate_triangle_z(triangle, angle)

    plot_triangles(triangle, rotated)

if __name__ == "__main__":
    main()
```

```
import math
import matplotlib.pyplot as plt

def rotate_point_y(x, y, z, angle_deg):
    """
    Rotate a point around the Y-axis by a given angle in degrees.
    """
    angle_rad = math.radians(angle_deg)
    cos_theta = math.cos(angle_rad)
    sin_theta = math.sin(angle_rad)

    x_new = x * cos_theta + z * sin_theta
    z_new = -x * sin_theta + z * cos_theta
    return (x_new, y, z_new)

def rotate_triangle_y(triangle, angle_deg):
    """
    Rotate all points of the triangle around the Y-axis.
    """
    return [rotate_point_y(x, y, z, angle_deg) for x, y, z in triangle]

def plot_triangles(original, rotated):
    """
    Plot the original and rotated triangles in 3D using matplotlib.
    """
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')

    x0, y0, z0 = zip(*original)
    x0 += (x0[0],)
    y0 += (y0[0],)
    z0 += (z0[0],)
    ax.plot(x0, y0, z0, color='blue', label='Original Triangle')
    ax.scatter(x0[:-1], y0[:-1], z0[:-1], color='blue')

    x1, y1, z1 = zip(*rotated)
    x1 += (x1[0],)
    y1 += (y1[0],)
    z1 += (z1[0],)
    ax.plot(x1, y1, z1, color='red', label='Rotated Triangle')
    ax.scatter(x1[:-1], y1[:-1], z1[:-1], color='red')

    ax.set_xlabel("X Axis")
    ax.set_ylabel("Y Axis")
    ax.set_zlabel("Z Axis")
    ax.set_title("Rotation of Triangle around Y-axis")
    ax.legend()
    ax.grid(True)
    plt.show()

def main():
    print("3D Triangle Rotation around Y-axis")

    triangle = []
    for i in range(3):
        print(f"\nEnter coordinates of point {i+1}:")
        x = float(input("  X: "))
        y = float(input("  Y: "))
        z = float(input("  Z: "))
        triangle.append((x, y, z))

    angle = float(input("\nEnter rotation angle (in degrees): "))
    rotated = rotate_triangle_y(triangle, angle)
    plot_triangles(triangle, rotated)

if __name__ == "__main__":
    main()
```

Scaling traingle 3d


```
import matplotlib.pyplot as plt

def scale_point(x, y, z, sx, sy, sz):
    """
    Scale a single 3D point by (sx, sy, sz).
    """
    return x * sx, y * sy, z * sz

def scale_triangle(triangle, sx, sy, sz):
    """
    Scale all vertices of the triangle.
    """
    return [scale_point(x, y, z, sx, sy, sz) for x, y, z in triangle]

def plot_triangles(original, scaled):
    """
    Plot the original and scaled triangles in 3D using matplotlib.
    """
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')

    x0, y0, z0 = zip(*original)
    x0 += (x0[0],)
    y0 += (y0[0],)
    z0 += (z0[0],)
    ax.plot(x0, y0, z0, color='blue', label='Original Triangle')
    ax.scatter(x0[:-1], y0[:-1], z0[:-1], color='blue')

    x1, y1, z1 = zip(*scaled)
    x1 += (x1[0],)
    y1 += (y1[0],)
    z1 += (z1[0],)
    ax.plot(x1, y1, z1, color='green', label='Scaled Triangle')
    ax.scatter(x1[:-1], y1[:-1], z1[:-1], color='green')

    ax.set_xlabel('X Axis')
    ax.set_ylabel('Y Axis')
    ax.set_zlabel('Z Axis')
    ax.set_title('3D Triangle Scaling')
    ax.legend()
    ax.grid(True)
    plt.show()

def main():
    print("3D Triangle Scaling Program")

    triangle = []
    for i in range(3):
        print(f"\nEnter coordinates of point {i+1}:")
        x = float(input("  X: "))
        y = float(input("  Y: "))
        z = float(input("  Z: "))
        triangle.append((x, y, z))

    print("\nEnter scaling factors:")
    sx = float(input("  Scale X by (sx): "))
    sy = float(input("  Scale Y by (sy): "))
    sz = float(input("  Scale Z by (sz): "))

    scaled = scale_triangle(triangle, sx, sy, sz)

    plot_triangles(triangle, scaled)

if __name__ == "__main__":
    main()
```

Translation 3d

```
import matplotlib.pyplot as plt

def translate_point_3d(x, y, z, tx, ty, tz):
    """
    Translate a 3D point by given translation values.
    """
    return x + tx, y + ty, z + tz

def plot_points(original, translated):
    """
    Plot the original and translated points in 3D using only matplotlib.
    """
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')


    x1, y1, z1 = original
    x2, y2, z2 = translated

    ax.scatter(x1, y1, z1, color='blue', s=50, label='Original Point')
    ax.text(x1, y1, z1, f'({x1}, {y1}, {z1})', color='blue')

    ax.scatter(x2, y2, z2, color='red', s=50, label='Translated Point')
    ax.text(x2, y2, z2, f'({x2}, {y2}, {z2})', color='red')

    ax.plot([x1, x2], [y1, y2], [z1, z2], linestyle='dashed', color='gray')

    ax.set_xlabel("X Axis")
    ax.set_ylabel("Y Axis")
    ax.set_zlabel("Z Axis")
    ax.set_title("3D Translation of a Point")
    ax.legend()
    ax.grid(True)

    plt.show()

def main():
    print("3D Point Translation and Plotting with Matplotlib")

    x = float(input("Enter original X coordinate: "))
    y = float(input("Enter original Y coordinate: "))
    z = float(input("Enter original Z coordinate: "))

    tx = float(input("Enter translation in X direction: "))
    ty = float(input("Enter translation in Y direction: "))
    tz = float(input("Enter translation in Z direction: "))

    original = (x, y, z)
    translated = translate_point_3d(x, y, z, tx, ty, tz)

    plot_points(original, translated)

if __name__ == "__main__":
    main()
```

Traingle translation 3d

```
import matplotlib.pyplot as plt

def translate_point(x, y, z, tx, ty, tz):
    """
    Translates a single 3D point by (tx, ty, tz).
    """
    return x + tx, y + ty, z + tz

def translate_triangle(triangle, tx, ty, tz):
    """
    Translates all vertices of the triangle.
    """
    return [translate_point(x, y, z, tx, ty, tz) for x, y, z in triangle]

def plot_triangles(original, translated):
    """
    Plots the original and translated triangles in 3D.
    """
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')

    x0, y0, z0 = zip(*original)
    x0 += (x0[0],)
    y0 += (y0[0],)
    z0 += (z0[0],)
    ax.plot(x0, y0, z0, color='blue', label='Original Triangle')
    ax.scatter(x0[:-1], y0[:-1], z0[:-1], color='blue')

    x1, y1, z1 = zip(*translated)
    x1 += (x1[0],)
    y1 += (y1[0],)
    z1 += (z1[0],)
    ax.plot(x1, y1, z1, color='red', label='Translated Triangle')
    ax.scatter(x1[:-1], y1[:-1], z1[:-1], color='red')

    ax.set_xlabel('X Axis')
    ax.set_ylabel('Y Axis')
    ax.set_zlabel('Z Axis')
    ax.set_title('3D Triangle Translation')
    ax.legend()
    ax.grid(True)
    plt.show()

def main():
    print("3D Triangle Translation Program")

    triangle = []
    for i in range(3):
        print(f"\nEnter coordinates of point {i+1}:")
        x = float(input("  X: "))
        y = float(input("  Y: "))
        z = float(input("  Z: "))
        triangle.append((x, y, z))

    print("\nEnter translation values:")
    tx = float(input("  Translate X by: "))
    ty = float(input("  Translate Y by: "))
    tz = float(input("  Translate Z by: "))

    translated = translate_triangle(triangle, tx, ty, tz)
    plot_triangles(triangle, translated)

if __name__ == "__main__":
    main()
```
